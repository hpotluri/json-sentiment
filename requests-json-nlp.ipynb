1. The following code accesses the [lyrics.ovh](https://lyricsovh.docs.apiary.io/#reference/0/lyrics-of-a-song/search) public api, searches for the lyrics of a song, and stores it in a dictionary object.  Write the resulting json to a file (either a JSON file or a pickle file; you choose). You will read in the contents of this file for future questions so we do not need to frequently access the API.

```python
import requests
import json

def get_song_lyrics(artist, title):
    url = f"https://api.lyrics.ovh/v1/{artist}/{title}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data["lyrics"]
    else:
        return None

def main():
    # Example artist and song title
    artist = "Coldplay"
    title = "Viva la Vida"

    lyrics = get_song_lyrics(artist, title)
    if lyrics:
        # Create a dictionary to store the lyrics
        lyrics_dict = {
            "artist": artist,
            "title": title,
            "lyrics": lyrics
        }
        
        # Write the dictionary to a JSON file
        with open("lyrics_data.json", "w") as json_file:
            json.dump(lyrics_dict, json_file)

if __name__ == "__main__":
    main()

```

3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

```python
import requests
import json

def get_and_save_lyrics(artist, title, filename):
    url = f"https://api.lyrics.ovh/v1/{artist}/{title}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        lyrics = data.get("lyrics", "")
        with open(filename, "w") as file:
            file.write(lyrics)
        print(f"Lyrics for {title} by {artist} saved to {filename}.")
    else:
        print(f"Failed to retrieve lyrics for {title} by {artist}.")

def main():
    # Test the function with four different songs
    get_and_save_lyrics("Coldplay", "Viva la Vida", "viva_la_vida_lyrics.txt")
    get_and_save_lyrics("Ed Sheeran", "Shape of You", "shape_of_you_lyrics.txt")
    get_and_save_lyrics("Adele", "Someone Like You", "someone_like_you_lyrics.txt")
    get_and_save_lyrics("Imagine Dragons", "Believer", "believer_lyrics.txt")

if __name__ == "__main__":
    main()
```
